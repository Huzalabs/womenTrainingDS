# Harvest Best Lawyers

```python
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup
import json
```

```python
from pydantic import BaseModel, Field, computed_field

class Lawyer(BaseModel):
    name: str = Field(title="Full Name", description="The full name of the lawyer")
    organisation: str = Field(title="Organisation", description="The organisation the lawyer works for")
    country: str = Field(title="Country", description="The country the lawyer is based in")
    practice_areas: list[str] = Field(title="Practice Areas", description="The practice areas the lawyer specialises in")
    specific_focus: str = Field(title="Specific Focus", description="The specific focus of the lawyer's practice")
    bio: str = Field(title="Biography", description="A brief biography of the lawyer")
    address_link: str = Field(title="Address", description="The address of the lawyer")
    organisation_link: str = Field(title="Organisation Link", description="The link to the organisation's website")

    @computed_field
    def address(self) -> str:
        return self.address_link.split('search/')[1].replace('+', ' ').replace('%2c', ',')
```

```python
# Read Countries
COUNTRIES = json.loads(open('countries.json').read())
COUNTRIES
```

```python
def process_profile(profile_html: str, country: str) -> Lawyer:
    soup = BeautifulSoup(profile_html, 'html.parser')
    slides = soup.find('div', class_='carousel-item')

    return Lawyer(
        name=soup.find('div', class_='card-title h6 mb-0').text.strip(),
        organisation=slides[0].find('a', class_='d-block py-1 text-charcoal td-none td-underline-hover firm-profile-link fw-800').text.strip(),
        country=country,
        practice_areas=[area.text.strip() for area in slides[0].find_all('div', class_='card-text fs-10 fw-600') if area.find_previous('div', class_='card-text fs-9 fw-800').text.strip() == 'Practice Areas'],
        specific_focus=slides[0].find('div', string='Specific Focus').find_next('div', class_='card-text fs-10 fw-600').text.strip(),
        bio=slides[1].find('p', class_='card-text fs-8 text-start text-truncate px-2').text.strip(),
        address_link=slides[2].find('a', class_=lambda x: x and 'bi-geo-alt-fill' in x),
        organisation_link=slides[0].find('a', class_='d-block py-1 text-charcoal td-none td-underline-hover firm-profile-link fw-800')
    )
```

```python
def get_profiles(practice_area_link: str, country: str) -> list[Lawyer]:
    content = requests.get(url='https://www.bestlawyers.com' + practice_area_link).content
    soup = BeautifulSoup(content, 'html.parser')

    profiles = soup.find_all('div', class_='card-scaler')

    return [process_profile(str(profile), country) for profile in profiles]
```

```python
def get_practice_areas(country: str) -> list[str]:
    content = requests.get(url=f'https://www.bestlawyers.com/{country}').content
    soup = BeautifulSoup(content, 'html.parser')

    # Find all the list of practice areas which are under the h2 with text as Practice Areas
    practice_areas = page.find('h2', string='Practice Areas').find_next('div').find_all('a')
    
    # Extract the href attribute from each of the practice areas and text into a dictionary
    practice_areas_dict = {area.text[1:-1].lower() : area['href'] for area in practice_areas}

    # replace all spaces i.e 1 or more with a single space including special characters
    return {re.sub(r'[\s\W]+', ' ', k.strip()).replace(' ', '-'): v for k, v in practice_areas_dict.items()}
```

```python
for country, link in COUNTRIES.items():
    practice_areas = get_practice_areas(link)
    for practice_area, practice_area_link in practice_areas.items():
        profiles = get_profiles(practice_area_link, country)
        DATABASE = DATABASE.append(pd.DataFrame(profiles), ignore_index=True)
```
